## In the following program we cluster the dataset n.1 with K-means.
From the plot of dataset n.1 we see 3 separated clusters. Thus k=3.

```python
import csv
import sys
from os.path import join

import numpy as np

def load_data(file_path, file_name):
   with open(join(file_path, file_name)) as csv_file:
       data_file = csv.reader(csv_file,delimiter=',')
       temp1 = next(data_file)
       n_samples = int(temp1[0])
       n_features = int(temp1[1])
       temp2 = next(data_file)
       feature_names = np.array(temp2[:n_features])

       data_list = [iter for iter in data_file]

       data = np.asarray(data_list, dtype=np.float64)

   return(data,feature_names,n_samples,n_features)

# The main program reads the input file containing the dataset
# file_path is the file path where the file with the data to be read are located
# we assume the file contains an example per line
# each example is a list of real values separated by a comma (csv format)
# The first line of the file contains the heading with:
# N_samples,n_features,
# The second line contains the feature names separated by a comma

#file_path="~/meo/Documents/Didattica/Laboratorio-15-16-Jupyter/"
file_path="./sample_data/"
# all the three datasets contain data points on (x,y)
file_name1="3-clusters.csv"

data1,feature_names1,n_samples1,n_features1 = load_data(file_path, file_name1)

from sklearn.cluster import KMeans

np.random.seed(5)

k=3
kmeans1 = KMeans(n_clusters=k, random_state=0).fit(data1)

i=0
for i in range(n_samples1):
    print("Example n."+str(i)+"=("+str(data1[i,0])+","+str(data1[i,1])+")")
    print("in cluster n."+str(kmeans1.labels_[i]))
```

In the following program we plot the clusters

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,8))

ax = fig.add_subplot(111)
fig.subplots_adjust(top=1)
ax.set_title('Clustered points in dataset n. 1')

ax.set_xlabel('x')
ax.set_ylabel('y')

# set the list of colors to be selected when plotting the different clusters
color=['b','g','r','c','m','y','k','w']

#plot the dataset
for clu in range(k):
    # collect the sequence of cooordinates of the points in each given cluster (determined by clu)
    data_list_x = [data1[i,0] for i in range(n_samples1) if kmeans1.labels_[i]==clu]
    data_list_y = [data1[i,1] for i in range(n_samples1) if kmeans1.labels_[i]==clu]
    plt.scatter(data_list_x, data_list_y, s=2, edgecolors='none', c=color[clu], alpha=0.5)

plt.show()
```

```python
import pandas as pd
import numpy as np
from os.path import join
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
```

```python
def loadDataset(filePath, fileName):
  dataSet = pd.read_csv(join(filePath, fileName))

  # get the header informations (number of elements and features) and print them
  header = dataSet.iloc[0]
  numElements = int(header.index[0])
  numFeatures = int(header.index[1])
  print("number of elemnts: " + str(numElements))
  print("number of features: " + str(numFeatures))

  # reset the columns' name of the dataframe and get the features' names
  new_header = dataSet.iloc[0] # grab the first row for the header
  dataSet = dataSet[1:] # take the data less the header row
  featureNames = new_header
  dataSet.columns = new_header # set the header row as the df header

  #convert the dataset into a numpy matrix
  data = np.asarray(dataSet.values, dtype=np.float64)

  return(data,featureNames,numElements,numFeatures)
```

```python
filePath="./sample_data/"
fileName2="CURE-complete.csv"
data2,featureNames2,numElements2,numFeatures2 = loadDataset(filePath, fileName2)

np.random.seed(5)
kmeans2 = KMeans(n_clusters=8, random_state=0).fit(data2)
```

```python
def plotDataset(indexDataset, dataset, numElements, kmeans):
  fig = plt.figure(figsize=(8,8))

  ax = fig.add_subplot(111)
  fig.subplots_adjust(top=1)
  ax.set_title('Clustered points in dataset n. ' + str(indexDataset))

  ax.set_xlabel('x')
  ax.set_ylabel('y')

  # set the list of colors to be selected when plotting the different clusters
  color=['b','g','r','c','m','y','k','tab:purple']

  #plot the dataset
  for clu in range(kmeans.n_clusters):
      # collect the sequence of cooordinates of the points in each given cluster (determined by clu)
      data_list_x = [dataset[i,0] for i in range(numElements) if kmeans.labels_[i]==clu]
      data_list_y = [dataset[i,1] for i in range(numElements) if kmeans.labels_[i]==clu]
      plt.scatter(data_list_x, data_list_y, s=2, edgecolors='none', c=color[clu], alpha=0.5)
      plt.scatter(kmeans.cluster_centers_[clu][0], kmeans.cluster_centers_[clu][1], edgecolors='black', color='black', alpha=0.5, marker="+") # print the center of the cluster

  plt.show()
```

```python
plotDataset(indexDataset=2, dataset=data2, numElements=numElements2, kmeans=kmeans2)
```

In the following cells I propose you to write a program that computes with a statistical measure of your choice a quantitative evaluation of clusters in the three datasets.

**Note:**
It is advisable to execute K-means a certain number of times (let us try 10 times) and then select the clustering solution that gives the best value of the evaluation measure.

```python
def computeSilhouettes(x,k):
  np.random.seed(5)
  kmeans = KMeans(n_clusters=k, n_init=10, random_state=0).fit(x)
  label = kmeans.labels_ # Clusters' labels of each point
  sil_coeff = silhouette_score(x, label, metric='euclidean')
  print("For n_clusters={}, The Silhouette Coefficient is {}".format(k, sil_coeff))

file_path="./sample_data/"
file_name1="3-clusters.csv"
data1,_,_,_ = loadDataset(file_path, file_name1)

filePath="./sample_data/"
fileName2="CURE-complete.csv"
data2,_,_,_ = loadDataset(filePath, fileName2)

filePath="./sample_data/"
fileName3="dataset-DBSCAN.csv"
data3,_,_,_ = loadDataset(filePath, fileName3)

print("first dataset:")
computeSilhouettes(data1,3)
print("")

print("second dataset:")
computeSilhouettes(data2,6)
print("")

print("third dataset:")
computeSilhouettes(data3,3)
```

As already done with classification by k-nn, plot in the cell below the quantitative measure of your choice (used above) with respect to an increasing value of k (the number of clusters) so that the best value of k can be selected.

```python
def computeSilhouettes(x):
  np.random.seed(5)
  sil_coeff = []
  for k in range(2,11):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=0).fit(x)
    label = kmeans.labels_ # Clusters' labels of each point
    sil_coeff.append(silhouette_score(x, label, metric='euclidean'))
  maxSilhouette = max(sil_coeff)
  indexMax = sil_coeff.index(maxSilhouette)
  print("For n_clusters={}, The max Silhouette Coefficient is {}".format(indexMax, maxSilhouette))
  return sil_coeff
```

```python
file_path="./sample_data/"
file_name1="3-clusters.csv"
data1,_,_,_ = loadDataset(file_path, file_name1)

filePath="./sample_data/"
fileName2="CURE-complete.csv"
data2,_,_,_ = loadDataset(filePath, fileName2)

filePath="./sample_data/"
fileName3="dataset-DBSCAN.csv"
data3,_,_,_ = loadDataset(filePath, fileName3)

print("first dataset:")
Silhouettes1 = computeSilhouettes(data1) # optimal k for the first dataset is k = 3 silhouetteMax: 0.7229402453727759
print("")

print("second dataset:")
Silhouettes2 = computeSilhouettes(data2) # optimal k for the second dataset is k = 6 silhouetteMax: 0.4282478787652604
print("")

print("third dataset:") # optimal k for the third dataset is k = 3 silhouetteMax: 0.47977770120112256
Silhouettes3 = computeSilhouettes(data3)

#plot
plt.plot(np.arange(2,11), Silhouettes1, label = "dataset 1")
plt.plot(np.arrange(2,11), Silhouettes2, label = "dataset 2")
plt.plot(np.arange(2,11), Silhouettes3, label = "dataset 3")
plt.legend()
plt.show()
```

```python
def plotDatasetDBSCAN(indexDataset, dataset, numElements, dbscan, feature_names):
  labels = dbscan.labels_

  # Number of clusters in labels, ignoring noise if present.
  n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
  n_noise_ = list(labels).count(-1)

  # print the statistic data
  print("Estimated number of clusters: %d" % n_clusters_)
  print("Estimated number of noise points: %d" % n_noise_)
  #print(f"Silhouette score: {silhouette_score(dataset, dbscan.labels_, metric='euclidean')}")

  fig = plt.figure(figsize=(8,8))

  ax = fig.add_subplot(111)
  fig.subplots_adjust(top=1)
  ax.set_title(f"Clustered points in dataset n. {indexDataset}")

  ax.set_xlabel(feature_names[0])
  ax.set_ylabel(feature_names[1])

  # set the list of colors to be selected when plotting the different clusters
  color=['b','g','r','c','m','y','k','w']

  n_clusters = len(set(dbscan.labels_))
  #plot the dataset
  for clu in range(n_clusters):
      # collect the sequence of cooordinates of the points in each given cluster (determined by clu)
      data_list_x = [dataset[i,0] for i in range(numElements) if dbscan.labels_[i]==clu]
      data_list_y = [dataset[i,1] for i in range(numElements) if dbscan.labels_[i]==clu]
      plt.scatter(data_list_x, data_list_y, s=2, edgecolors='none', cmap="rainbow", alpha=0.5)

  plt.show()
```

```python
# we choose the dataset 2
filePath="./sample_data/"
fileName2="CURE-complete.csv"
data2,featureNames2,numElements2,numFeatures2 = loadDataset(filePath, fileName2)
dbscan = DBSCAN(eps=4.5, min_samples=50).fit(data2)
plotDatasetDBSCAN(indexDataset=2, dataset=data2, numElements=numElements2, dbscan=dbscan, feature_names=featureNames2)
```

```python
# we choose minPts = 100
# create and fit the knn algorithm, then get the distances of the 100-th nearest neighbour of the points
from kneed import KneeLocator
neighbors = NearestNeighbors(n_neighbors=100)
neighbors_fit = neighbors.fit(data2) # knn (k=100)
distances, indices = neighbors_fit.kneighbors(data2)

distances = [distances[i][100 - 1] for i in range(len(distances))]
distances = np.sort(distances, axis=0)
plt.plot(distances)

kn = KneeLocator(x=distances, y=distances, curve='convex', direction='increasing')
print(f"Il gomito si trova a: {kn.knee}") # takes the starting value of the elbow
```
